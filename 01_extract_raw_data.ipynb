```python
import zipfile
from pathlib import Path

# ======================================================
# 1. 确保当前工作目录是项目根目录
#    如果你在 VS Code 中打开了 dshw-p02b 文件夹，
#    那么当前工作目录已经是项目根目录，这步不会报错。
#    如果不是，可以手动切换（取消下面一行的注释并修改路径）：
# ======================================================
# import os
# os.chdir("/Users/tutu/Desktop/dshw-p02b")

# 验证当前工作目录
print(f"当前工作目录: {Path.cwd()}")

# ======================================================
# 2. 定义路径（相对路径，适用于项目根目录下运行）
# ======================================================
zip_dir = Path("data/data_raw_zip")
raw_dir = Path("data/raw")

# 检查 zip_dir 是否存在
if not zip_dir.exists():
    print(f"❌ 错误：找不到 {zip_dir} 文件夹！请确认路径和当前工作目录。")
    # 如果找不到，列出当前目录内容帮助排查
    print("当前目录内容：", list(Path.cwd().iterdir()))
else:
    # 创建 raw 文件夹（如果不存在）
    raw_dir.mkdir(parents=True, exist_ok=True)

    # 获取所有 .zip 文件（不区分大小写）
    zip_files = sorted([f for f in zip_dir.glob("*") if f.suffix.lower() == '.zip'])

    print(f"在 {zip_dir} 中找到 {len(zip_files)} 个压缩包：")
    for f in zip_files:
        print(f"  - {f.name}")

    if len(zip_files) == 0:
        print("⚠️ 没找到任何 .zip 文件，请检查文件名是否以 .zip 结尾。")
    else:
        # 逐个解压
        for zf in zip_files:
            try:
                with zipfile.ZipFile(zf, 'r') as z:
                    # 如果压缩包内的文件名有乱码（中文编码问题），可以尝试指定编码
                    # 大多数 CSMAR 压缩包是 GBK 编码，这里先按默认处理，如有乱码再调整
                    z.extractall(raw_dir)
                print(f"✔ 已解压: {zf.name}")
            except Exception as e:
                print(f"✘ 解压失败 {zf.name}: {e}")

        print(f"\n✅ 全部解压完成！解压后的文件在: {raw_dir}")
```

```python
import pandas as pd
from pathlib import Path

raw_path = Path("data/raw")
for file in sorted(raw_path.glob("*")):
    if file.suffix.lower() in ['.csv', '.txt']:
        try:
            df = pd.read_csv(file, nrows=0, encoding='gbk')
        except:
            df = pd.read_csv(file, nrows=0, encoding='utf-8')
    elif file.suffix.lower() in ['.xlsx', '.xls']:
        df = pd.read_excel(file, nrows=0)
    else:
        continue
    print(f"\n===== {file.name} =====")
    print(df.columns.tolist())
```
